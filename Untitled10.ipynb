```python
# ========= 0) 准备 =========
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler   # 或 MinMaxScaler（老师建议二者都试）
from sklearn.decomposition import PCA

plt.style.use("ggplot")

# ========= 1) 数据清理（NaN/Inf） =========
# wide_2023_clean: 你的宽表（例如 2023 年，非金融公司）
# 如有元数据列（Ticker/Year/Industry/TickerYear），按需放入 meta_cols 排除
meta_cols = [c for c in ["Ticker","Year","Industry","TickerYear"] if c in wide_2023_clean.columns]

# 只对数值特征做处理
feat_cols = [c for c in wide_2023_clean.columns if c not in meta_cols]
X_raw = wide_2023_clean[feat_cols].copy()

# 把 inf/-inf → NaN，再用列内中位数填补（比补 0 更稳）
X_num = X_raw.replace([np.inf, -np.inf], np.nan)
X_num = X_num.apply(lambda s: s.fillna(s.median()), axis=0)

# ========= 2) 标准化（Normalization / Scaling） =========
# 标准化方案一：StandardScaler（Z-score）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num.values)

# 生成标准化后的 DataFrame（统一变量名：z_df）
z_df = pd.DataFrame(X_scaled, index=wide_2023_clean.index, columns=feat_cols)
# 把元数据列（若有）拼回去，便于后续着色/标注
for c in meta_cols:
    z_df[c] = wide_2023_clean[c].values

print("z_df shape:", z_df.shape)
display(z_df.head())

# ========= 3) PCA 拟合（降维） =========
X = z_df[feat_cols].values
pca = PCA(n_components=3, random_state=0)  # 3个主成分足够画 Scree 和 2D/3D
Z = pca.fit_transform(X)                   # Z: PCA scores

evr = pca.explained_variance_ratio_
cum = np.cumsum(evr)
print("Explained variance ratio:", np.round(evr, 4))
print("Cumulative EVR:", np.round(cum, 4))

# ========= 4) Scree Plot（累计解释方差） =========
plt.figure(figsize=(6,4))
plt.plot(range(1, len(evr)+1), cum, marker="o")
plt.axhline(0.80, ls="--", lw=1)  # 80% 参考线
plt.xlabel("Number of PCs")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Scree (2023)")
plt.tight_layout()
plt.show()

# ========= 5) 2D Scores Plot（PC1 vs PC2） =========
plt.figure(figsize=(7,6))

# 如果有 Industry/Sector，就按行业上色；否则统一颜色
if "Industry" in z_df.columns:
    sectors = z_df["Industry"].astype(str)
    uniq = sectors.unique()
    colors = dict(zip(uniq, plt.rcParams['axes.prop_cycle'].by_key()['color'][:len(uniq)]))
    for sec in uniq:
        idx = (sectors == sec).values
        plt.scatter(Z[idx,0], Z[idx,1], s=28, alpha=0.85, label=sec, c=colors[sec])
    plt.legend(title="Industry", fontsize=8)
else:
    plt.scatter(Z[:,0], Z[:,1], s=28, alpha=0.85)

# 可选：给部分点加 Ticker 标签（避免太拥挤）
if "Ticker" in z_df.columns:
    for i, tic in enumerate(z_df["Ticker"].astype(str).values):
        if i % 8 == 0:
            plt.text(Z[i,0], Z[i,1], tic, fontsize=7, ha="center", va="bottom")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA on Financial Ratios (2023)")
plt.tight_layout()
plt.show()

# ========= 6) Loadings（解释 PC1/PC2 是什么） =========
# components_.T: 行=特征，列=PC
loadings = pd.DataFrame(
    pca.components_.T,
    index=feat_cols,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)

top_pc1 = loadings["PC1"].abs().sort_values(ascending=False).head(10)
top_pc2 = loadings["PC2"].abs().sort_values(ascending=False).head(10)

print("\nPC1 最主要的 10 个指标（按 |loading| 排序）：")
display(top_pc1.to_frame("abs_loading_PC1"))

print("\nPC2 最主要的 10 个指标（按 |loading| 排序）：")
display(top_pc2.to_frame("abs_loading_PC2"))
```

## Umap
“在对比 t-SNE 与 UMAP 的结果时发现，t-SNE 更容易形成紧密簇，而 UMAP 的点分布相对稀疏。这与两者的优化目标有关：t-SNE 注重保持局部邻居关系，而 UMAP 同时尝试保留全局拓扑，因此在小样本下更容易出现点的分散。尽管如此，UMAP 结果仍显示部分行业公司在二维空间上具有相对一致的分布，这支持了财务比率在行业分类中的潜在作用。”
PCA：解释率有限 → 可以写成“线性方法不足”。

t-SNE：行业簇比较明显 → 写成“非线性方法更能揭示结构”。

UMAP：比较分散 → 写成“更强调全局，但在小样本下效果一般”。

```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, adjusted_rand_score, normalized_mutual_info_score, f1_score
from scipy.optimize import linear_sum_assignment

def evaluate_and_align(y_true, y_pred, name="Clustering"):
    cm = confusion_matrix(y_true, y_pred)
    r, c = linear_sum_assignment(-cm)  # 匈牙利算法对齐簇标签
    mapping = {c_i: r_i for r_i, c_i in zip(r, c)}
    y_pred_m = np.array([mapping.get(v, -1) for v in y_pred])

    acc = accuracy_score(y_true, y_pred_m)
    ari = adjusted_rand_score(y_true, y_pred_m)
    nmi = normalized_mutual_info_score(y_true, y_pred_m)
    f1  = f1_score(y_true, y_pred_m, average="macro")
    print(f"{name} -> ACC: {acc:.3f}  ARI: {ari:.3f}  NMI: {nmi:.3f}  F1: {f1:.3f}")
    return y_pred_m
```

```python
# ==== Prep: encode true labels as numbers ====
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# y_true（字符串）→ 数字 0..4
y_true_str = np.array([sector_map[t] for t in Z_df.index])
le = LabelEncoder()
y_true = le.fit_transform(y_true_str)   # 数字标签，供评估函数使用

# ==== K-Means ====
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=len(le.classes_), random_state=42)
labels_km = kmeans.fit_predict(Z_df.values)

# ==== 对齐 + 打分（evaluate_and_align 用数字更稳）====
labels_km_m = evaluate_and_align(y_true, labels_km, "K-Means")

# ==== 结果表（把数字映回行业名，便于查看）====
true_names = le.inverse_transform(y_true)
pred_names = le.inverse_transform(labels_km_m)

km_out = pd.DataFrame({
    "Ticker": Z_df.index,
    "True": true_names,
    "KMeans": pred_names
})
km_out["Correct?"] = km_out["True"] == km_out["KMeans"]
km_out.head(10)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def pretty_scatter(X2, labels, title, annotate=True, fname=None):
    """X2: (n,2) 数组或DataFrame；labels: 行业或簇标签（与行对齐）"""
    XY = pd.DataFrame(X2, index=labels.index if hasattr(labels,'index') else None, columns=['x','y'])
    secs = pd.Series(labels).astype(str)
    palette = dict(zip(secs.unique(), plt.cm.tab10.colors[:len(secs.unique())]))
    plt.figure(figsize=(6.6,5.6))
    for s in secs.unique():
        m = (secs==s)
        plt.scatter(XY.loc[m,'x'], XY.loc[m,'y'], s=36, alpha=0.85, edgecolors='white', linewidths=0.4,
                    c=[palette[s]], label=s)
    if annotate and len(XY)<=120:
        for t,(x0,y0) in XY.iterrows(): plt.text(x0,y0,str(t),fontsize=7)
    plt.title(title); plt.xlabel('dim1'); plt.ylabel('dim2'); plt.legend(title='', bbox_to_anchor=(1.02,1), loc='upper left')
    plt.tight_layout()
    if fname: plt.savefig(fname, dpi=200)
    plt.show()
```

```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

# 1) 用 PCA 把高维特征投到 2D（只为可视化）
pca2 = PCA(n_components=2, random_state=42)
XY = pca2.fit_transform(Z_df.values)                    # (n,2) numpy
XY_df = pd.DataFrame(XY, index=Z_df.index, columns=["x","y"])

# 2) 聚类标签（来自你在高维上的 KMeans）
labels = pd.Series((best_km.labels_ if 'best_km' in globals() else km.labels_),
                   index=Z_df.index, name="Cluster")
k_list = sorted(labels.unique())
n_clusters = len(k_list)

# 3) 在 2D 上用 KNN 拟合这些簇标签（仅为了画边界）
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(XY, labels.values)

# 4) 画决策边界（背景色）
x_min, x_max = XY[:,0].min()-0.6, XY[:,0].max()+0.6
y_min, y_max = XY[:,1].min()-0.6, XY[:,1].max()+0.6
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 400),
                     np.linspace(y_min, y_max, 400))
zz = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(7,6))
plt.contourf(xx, yy, zz, alpha=0.15, levels=np.arange(-0.5, n_clusters+0.5, 1),
             cmap='tab10')

# 5) 画散点 + 质心（X）
palette = dict(zip(k_list, plt.cm.tab10.colors[:n_clusters]))
for k in k_list:
    pts = XY[labels.values==k]
    plt.scatter(pts[:,0], pts[:,1], s=46, alpha=0.9,
                edgecolors='white', linewidths=0.6,
                c=[palette[k]], label=f"Cluster {k}")
    c = pts.mean(axis=0)                                  # 质心（在2D）
    plt.scatter(c[0], c[1], s=180, marker="X",
                edgecolors="black", linewidths=1.0,
                facecolors=palette[k])

plt.title(f"K-Means (k={n_clusters}) — drawn in PCA 2D")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.legend(title="", bbox_to_anchor=(1.02,1), loc="upper left")
plt.tight_layout(); plt.show()
```
